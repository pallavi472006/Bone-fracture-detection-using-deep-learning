```python
# ===============================
# 📌 Install & Import Libraries
# ===============================
!pip install opencv-python tensorflow

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout

# ===============================
# 📌 Mount Google Drive
# ===============================
from google.colab import drive
drive.mount('/content/drive')

# ===============================
# 📌 Dataset Paths
# ===============================
folder_path = '/content/drive/My Drive/FracNet'
fracture_path = os.path.join(folder_path, 'Fractured')
not_fracture_path = os.path.join(folder_path, 'Not Fractured')

# ===============================
# 📌 Count Images
# ===============================
total_fracture = len(os.listdir(fracture_path))
total_not_fracture = len(os.listdir(not_fracture_path))
print('Total Fractured Images:', total_fracture)
print('Total Not Fractured Images:', total_not_fracture)

# ===============================
# 📌 Load Images
# ===============================
IMG_SIZE = 256
MAX_IMAGES = 2000  # safety cap

X_data = []
Y_labels = []

# Fractured = 0
for i, filename in enumerate(os.listdir(fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        X_data.append(img)
        Y_labels.append(0)

print("✅ Loaded Fractured images")

# Not Fractured = 1
for i, filename in enumerate(os.listdir(not_fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(not_fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        X_data.append(img)
        Y_labels.append(1)

print("✅ Loaded Not Fractured images")

# Convert to numpy arrays
X_data = np.array(X_data, dtype="float32") / 255.0
Y_labels = np.array(Y_labels)

print("📊 Final Dataset Shape:")
print("X_data:", X_data.shape)
print("Y_labels:", Y_labels.shape)

# ===============================
# 📌 Train-Test Split
# ===============================
Y_labels = tf.keras.utils.to_categorical(Y_labels, num_classes=2)

x_train, x_test, y_train, y_test = train_test_split(
    X_data, Y_labels, test_size=0.2, random_state=42, stratify=Y_labels
)

print("X_train shape:", x_train.shape)
print("Y_train shape:", y_train.shape)
print("X_test shape:", x_test.shape)
print("Y_test shape:", y_test.shape)

# ===============================
# 📌 Build CNN Model
# ===============================
print("🔨 Building CNN Model...")

model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3), activation="relu", input_shape=(IMG_SIZE, IMG_SIZE, 3)))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=2, activation="softmax"))

print("✅ Model Summary")
model.summary()

# ===============================
# 📌 Compile Model
# ===============================
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])

# ===============================
# 📌 Train Model (with Validation)
# ===============================
epoch = 10
batch_size = 64

history = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epoch,
    verbose=1,
    validation_data=(x_test, y_test)
)

print("✅ Training Complete!")

# ===============================
# 📊 Plot Training & Validation Graphs
# ===============================
plt.figure(figsize=(12, 5))

# Loss graph
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="Training Loss", marker='o')
plt.plot(history.history['val_loss'], label="Validation Loss", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()

# Accuracy graph
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="Training Accuracy", marker='o')
plt.plot(history.history['val_accuracy'], label="Validation Accuracy", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()

plt.show()

# ===============================
# 📌 Confusion Matrix
# ===============================
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fractured", "Not Fractured"])
disp.plot(cmap=plt.cm.Blues, values_format="d")
plt.title("Confusion Matrix")
plt.show()

# ===============================
# 📌 Print Final Validation Accuracy & Loss
# ===============================
final_val_acc = history.history['val_accuracy'][-1]
final_val_loss = history.history['val_loss'][-1]

print(f"📊 Final Validation Accuracy: {final_val_acc:.4f}")
print(f"📉 Final Validation Loss: {final_val_loss:.4f}")


# ===============================
# 📋 Classification Report
# ===============================
from sklearn.metrics import classification_report

# Generate classification report
report = classification_report(y_true, y_pred_classes, target_names=["Fractured", "Not Fractured"])
print("📋 Classification Report:\n")
print(report)


```
