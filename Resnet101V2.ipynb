```python
# ===============================
# ğŸ“¦ Install Required Libraries (Run Once)
# ===============================
# !pip install opencv-python tensorflow scikit-learn matplotlib

# ===============================
# ğŸ“š Import Libraries
# ===============================
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import ResNet101V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input

# ===============================
# ğŸ“‚ Dataset Paths
# ===============================
folder_path = r"C:\Users\palla\OneDrive\Desktop\FracNet"  # Update this path
fracture_path = os.path.join(folder_path, 'Fractured')
not_fracture_path = os.path.join(folder_path, 'Not Fractured')

print('Total Fractured Images:', len(os.listdir(fracture_path)))
print('Total Not Fractured Images:', len(os.listdir(not_fracture_path)))

# ===============================
# ğŸ§  Load & Preprocess Images
# ===============================
IMG_SIZE = 224
MAX_IMAGES = 2000

X_data, Y_labels = [], []

# Fractured = 0
for i, filename in enumerate(os.listdir(fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = preprocess_input(img)
        X_data.append(img)
        Y_labels.append(0)

# Not Fractured = 1
for i, filename in enumerate(os.listdir(not_fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(not_fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = preprocess_input(img)
        X_data.append(img)
        Y_labels.append(1)

X_data = np.array(X_data, dtype="float32")
Y_labels = tf.keras.utils.to_categorical(Y_labels, num_classes=2)

print("ğŸ“Š Final Dataset Shape:")
print("X_data:", X_data.shape)
print("Y_labels:", Y_labels.shape)

# ===============================
# âœ‚ï¸ Train-Test Split
# ===============================
x_train, x_test, y_train, y_test = train_test_split(
    X_data, Y_labels, test_size=0.2, random_state=42, stratify=Y_labels
)

# ===============================
# ğŸ§¹ Remove Duplicate Images Between Train & Test
# ===============================
train_bytes = [img.tobytes() for img in x_train]
test_bytes = [img.tobytes() for img in x_test]

duplicate_indices = [i for i, img in enumerate(test_bytes) if img in train_bytes]

if duplicate_indices:
    print(f"âš ï¸ Found {len(duplicate_indices)} duplicate images â€” removing from test set.")
    x_test = np.delete(x_test, duplicate_indices, axis=0)
    y_test = np.delete(y_test, duplicate_indices, axis=0)
else:
    print("âœ… No duplicate images found between train and test.")

print("Train set:", x_train.shape, y_train.shape)
print("Test set:", x_test.shape, y_test.shape)

# ===============================
# ğŸ§© Load ResNet101V2 Base Model
# ===============================
print("ğŸ”¨ Building ResNet101V2 Model...")

base_model = ResNet101V2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = False  # Freeze for transfer learning

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)
predictions = Dense(2, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

print("âœ… Model Summary")
model.summary()

# ===============================
# âš™ï¸ Compile Model
# ===============================
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# ===============================
# ğŸš€ Train Model
# ===============================
epochs = 15
batch_size = 32

history = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(x_test, y_test),
    verbose=1
)

# ===============================
# ğŸ“ˆ Plot Training & Validation Graphs
# ===============================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="Training Loss", marker='o')
plt.plot(history.history['val_loss'], label="Validation Loss", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="Training Accuracy", marker='o')
plt.plot(history.history['val_accuracy'], label="Validation Accuracy", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()

plt.show()

# ===============================
# ğŸ“Š Confusion Matrix
# ===============================
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fractured", "Not Fractured"])
disp.plot(cmap=plt.cm.Blues, values_format="d")
plt.title("Confusion Matrix")
plt.show()

# ===============================
# ğŸ“‹ Final Validation Metrics
# ===============================
final_val_acc = history.history['val_accuracy'][-1]
final_val_loss = history.history['val_loss'][-1]
print(f"ğŸ“Š Final Validation Accuracy: {final_val_acc:.4f}")
print(f"ğŸ“‰ Final Validation Loss: {final_val_loss:.4f}")

# ===============================
# ğŸ§¾ Classification Report
# ===============================
report = classification_report(y_true, y_pred_classes, target_names=["Fractured", "Not Fractured"])
print("ğŸ“‹ Classification Report:\n")
print(report)
```
