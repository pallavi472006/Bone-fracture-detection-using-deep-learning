```python
## 🧠 Fracture Detection with MobileNetV2 (Jupyter Version)

# ===============================
# 📌 Install Required Libraries (Run Once)
# ===============================
# Uncomment and run this cell once if needed
# !pip install opencv-python tensorflow scikit-learn matplotlib

# ===============================
# 📌 Import Libraries
# ===============================
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

# ===============================
# 📌 Dataset Paths
# ===============================
# Update this path to your local dataset directory
folder_path = r'C:\Users\palla\OneDrive\Desktop\FracNet'  # Example path
fracture_path = os.path.join(folder_path, 'Fractured')
not_fracture_path = os.path.join(folder_path, 'Not Fractured')

# ===============================
# 📌 Count Images
# ===============================
print('Total Fractured Images:', len(os.listdir(fracture_path)))
print('Total Not Fractured Images:', len(os.listdir(not_fracture_path)))

# ===============================
# 📌 Load & Preprocess Images
# ===============================
IMG_SIZE = 224
MAX_IMAGES = 2000

X_data = []
Y_labels = []

# Fractured = 0
for i, filename in enumerate(os.listdir(fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = preprocess_input(img)
        X_data.append(img)
        Y_labels.append(0)

# Not Fractured = 1
for i, filename in enumerate(os.listdir(not_fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(not_fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = preprocess_input(img)
        X_data.append(img)
        Y_labels.append(1)

X_data = np.array(X_data, dtype="float32")
Y_labels = tf.keras.utils.to_categorical(Y_labels, num_classes=2)

print("📊 Final Dataset Shape:")
print("X_data:", X_data.shape)
print("Y_labels:", Y_labels.shape)

# ===============================
# 📌 Train-Test Split
# ===============================
x_train, x_test, y_train, y_test = train_test_split(
    X_data, Y_labels, test_size=0.2, random_state=42, stratify=Y_labels
)

# ===============================
# 🧹 Remove Duplicate Images Between Train & Test
# ===============================
train_bytes = [img.tobytes() for img in x_train]
test_bytes = [img.tobytes() for img in x_test]

duplicate_indices = [i for i, img in enumerate(test_bytes) if img in train_bytes]

if duplicate_indices:
    print(f"⚠️ Found {len(duplicate_indices)} duplicate images — removing from test set.")
    x_test = np.delete(x_test, duplicate_indices, axis=0)
    y_test = np.delete(y_test, duplicate_indices, axis=0)
else:
    print("✅ No duplicate images found between train and test.")

print("Train set:", x_train.shape, y_train.shape)
print("Test set:", x_test.shape, y_test.shape)

# Verify cleanup
train_set = set([img.tobytes() for img in x_train])
test_set = set([img.tobytes() for img in x_test])
duplicates = train_set.intersection(test_set)
print(f"⚡ Duplicate images between train/test after cleaning: {len(duplicates)}")

# ===============================
# 📌 Load MobileNetV2 Base Model
# ===============================
print("🔨 Building MobileNetV2 Model...")

base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(2, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

print("✅ Model Summary")
model.summary()

# ===============================
# 📌 Compile Model
# ===============================
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# ===============================
# 📌 Train Model
# ===============================
epochs = 10
batch_size = 64

history = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(x_test, y_test),
    verbose=1
)

# ===============================
# 📊 Plot Training & Validation Graphs
# ===============================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="Training Loss", marker='o')
plt.plot(history.history['val_loss'], label="Validation Loss", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="Training Accuracy", marker='o')
plt.plot(history.history['val_accuracy'], label="Validation Accuracy", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()

plt.show()

# ===============================
# 📌 Confusion Matrix
# ===============================
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fractured", "Not Fractured"])
disp.plot(cmap=plt.cm.Blues, values_format="d")
plt.title("Confusion Matrix")
plt.show()

# ===============================
# 📌 Final Validation Metrics
# ===============================
final_val_acc = history.history['val_accuracy'][-1]
final_val_loss = history.history['val_loss'][-1]
print(f"📊 Final Validation Accuracy: {final_val_acc:.4f}")
print(f"📉 Final Validation Loss: {final_val_loss:.4f}")

# ===============================
# 📋 Classification Report
# ===============================
report = classification_report(y_true, y_pred_classes, target_names=["Fractured", "Not Fractured"])
print("📋 Classification Report:\n")
print(report)
```
