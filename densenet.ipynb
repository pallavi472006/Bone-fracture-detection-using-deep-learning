```python
# ===============================
# 📌 Install & Import Libraries
# ===============================
!pip install opencv-python tensorflow

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical

# ===============================
# 📌 Mount Google Drive
# ===============================
from google.colab import drive
drive.mount('/content/drive')

# ===============================
# 📌 Dataset Paths
# ===============================
folder_path = '/content/drive/My Drive/FracNet'
fracture_path = os.path.join(folder_path, 'Fractured')
not_fracture_path = os.path.join(folder_path, 'Not Fractured')

# ===============================
# 📌 Count Images
# ===============================
total_fracture = len(os.listdir(fracture_path))
total_not_fracture = len(os.listdir(not_fracture_path))
print('Total Fractured Images:', total_fracture)
print('Total Not Fractured Images:', total_not_fracture)

# ===============================
# 📌 Load Images
# ===============================
IMG_SIZE = 224  # DenseNet requires 224x224
MAX_IMAGES = 2000  # optional limit

X_data = []
Y_labels = []

# Fractured = 0
for i, filename in enumerate(os.listdir(fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        X_data.append(img)
        Y_labels.append(0)

print("✅ Loaded Fractured images")

# Not Fractured = 1
for i, filename in enumerate(os.listdir(not_fracture_path)):
    if i >= MAX_IMAGES: break
    path = os.path.join(not_fracture_path, filename)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        X_data.append(img)
        Y_labels.append(1)

print("✅ Loaded Not Fractured images")

# Convert to numpy arrays
X_data = np.array(X_data, dtype="float32") / 255.0
Y_labels = to_categorical(Y_labels, num_classes=2)

print("📊 Final Dataset Shape:")
print("X_data:", X_data.shape)
print("Y_labels:", Y_labels.shape)

# ===============================
# 📌 Train-Test Split
# ===============================
x_train, x_test, y_train, y_test = train_test_split(
    X_data, Y_labels, test_size=0.2, random_state=42, stratify=Y_labels
)

# ===============================
# 📌 Build DenseNet Model
# ===============================
print("🔨 Building DenseNet Model...")

base_model = DenseNet121(include_top=False, weights='imagenet', input_tensor=Input(shape=(IMG_SIZE, IMG_SIZE, 3)))

# Freeze the base model
base_model.trainable = False

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(2, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

print("✅ Model Summary")
model.summary()

# ===============================
# 📌 Compile Model
# ===============================
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# ===============================
# 📌 Train Model (with Validation)
# ===============================
epochs = 10
batch_size = 32

history = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(x_test, y_test),
    verbose=1
)

print("✅ Training Complete!")

# ===============================
# 📊 Plot Training & Validation Graphs
# ===============================
plt.figure(figsize=(12, 5))

# Loss graph
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="Training Loss", marker='o')
plt.plot(history.history['val_loss'], label="Validation Loss", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()

# Accuracy graph
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="Training Accuracy", marker='o')
plt.plot(history.history['val_accuracy'], label="Validation Accuracy", marker='o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()

plt.show()

# ===============================
# 📌 Confusion Matrix
# ===============================
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fractured", "Not Fractured"])
disp.plot(cmap=plt.cm.Blues, values_format="d")
plt.title("Confusion Matrix")
plt.show()

# ===============================
# 📌 Final Validation Accuracy & Loss
# ===============================
final_val_acc = history.history['val_accuracy'][-1]
final_val_loss = history.history['val_loss'][-1]

print(f"📊 Final Validation Accuracy: {final_val_acc:.4f}")
print(f"📉 Final Validation Loss: {final_val_loss:.4f}")

# ===============================
# 📋 Classification Report
# ===============================
report = classification_report(y_true, y_pred_classes, target_names=["Fractured", "Not Fractured"])
print("📋 Classification Report:\n")
print(report)
```
